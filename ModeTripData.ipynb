```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the data into a dataframe
data = pd.read_csv(r'E:\Trip_data\MODE PREF.csv')

# Define the mapping between the strings and numerical values
mapping = {
    "Not At All Important": 1,
    "Not Important": 2,
    "Immertirial": 3,
    "Important": 4,
    "Very Important": 5
}

# Define the columns to convert
columns = ['ToDropMyChildren', 'Shopping', 'TravelTime', 'TravelCost', 'CongestionAndPollution',
           'Privacy', 'Comfort', 'Saftey', 'FrequencyinPublicTransport', 'FeederDistance',
           'ClimaticConditions', 'TravellingAllowance', 'FuelCost', 'ParkingDifficultyAndParkingCost',
           'NatureofWork', 'PhysicalConstraints']

# Apply the mapping to the selected columns
data[columns] = data[columns].replace(mapping)

# Create a heatmap plot
plt.figure(figsize=(10,8))
sns.heatmap(data[columns], cmap='coolwarm', annot=True)
plt.show()
```
