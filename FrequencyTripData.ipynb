```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data into a dataframe
fd = pd.read_csv(r'E:\Trip_data\FREQUENCY.csv')

# Define the mapping between the strings and numerical values
mapping = {
    "Not Applicable": 1,
    "Daily": 2,
    "Once In A Week": 3,
    "Once In Two Weeks": 4,
    "Once In A Month": 5
}

# Define the columns to convert
columns = ['Work', 'Educational', 'Shopping', 'Religious', 'Hospitals', 'Bank/ATM',
           'PostOffices', 'Threaters', 'Parks', 'RailwayStations', 'Markets',
           'PetrolPumps', 'Resturants', 'Saloons', 'FamilyVisits',
           'Return Home', 'OtherFreq']

# Apply the mapping to the selected columns
fd[columns] = fd[columns].replace(mapping)

# Create a heatmap plot
plt.figure(figsize=(20,14))
sns.heatmap(fd[columns], cmap='coolwarm', annot=True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# read the CSV file
df = pd.read_csv('E:\Trip_data\FREQUENCY.csv')

# map the values to numbers
values = {
    "Not Applicable": 1,
    "Daily": 2,
    "Once In A Week": 3,
    "Once In Two Weeks": 4,
    "Once In A Month": 5
}
df.replace(values, inplace=True)

# select the columns to plot
columns = ['Work', 'Educational', 'Shopping', 'Religious', 'Hospitals', 'Bank/ATM', 'PostOffices', 'Threaters', 'Parks', 'RailwayStations', 'Markets', 'PetrolPumps', 'Resturants', 'Saloons', 'FamilyVisits', 'Return Home', 'OtherFreq']
data = df[columns]

# plot the heatmap
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
```
