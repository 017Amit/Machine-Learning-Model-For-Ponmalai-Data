```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from CSV file
data = pd.read_csv('E:\\Trip_data\\MODE PREF.csv')

# Create a list of columns to plot
columns_to_plot = ['ToDropMyChildren', 'Shopping', 'TravelTime', 'TravelCost', 'CongestionAndPollution', 
                   'Privacy', 'Comfort', 'Saftey', 'FrequencyinPublicTransport', 'FeederDistance', 
                   'ClimaticConditions', 'TravellingAllowance', 'FuelCost', 'ParkingDifficultyAndParkingCost',
                   'NatureofWork', 'PhysicalConstraints']

# Convert the string values to numeric values using a dictionary
mapping = {'Not At All Important': 1, 'Not Important': 2, 'Immertirial': 3, 'Important': 4, 'Very Important': 5}
data[columns_to_plot] = data[columns_to_plot].replace(mapping)

# Create a custom color palette
palette = sns.diverging_palette(220, 20, as_cmap=True)

# Create a heatmap of the data
fig, ax = plt.subplots(figsize=(15, 8))
sns.heatmap(data[columns_to_plot].corr(), cmap=palette, annot=True, fmt='.2f', ax=ax, cbar_kws={'label': 'Correlation Coefficient'})
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=12)

# Set the axis labels and title
ax.set_xlabel('Factors Influencing Mode of Transport Decision', fontsize=16, color = "blue")
ax.set_ylabel('Factors Influencing Mode of Transport Decision', fontsize=16, color = "blue")
ax.set_title('Heatmap of Correlation Matrix for Factors Influencing Mode of Transport Decision', fontsize=18, color = "red")

# Set the font size of the annotations
for text in ax.texts:
    text.set_fontsize(12)

# Show the plot
plt.show()
```
